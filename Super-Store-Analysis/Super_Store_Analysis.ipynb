## Exploratory Data Analysis

### 1. Which Category is Best Selling and Most Profitable?

To find the best-selling and most profitable categories, we will group the sales, profit, and quantity of products by category. We will then use the sum() function to find the total sales, profit, and quantity for each category. Finally, we will use the DataFrame() function to create a table of the results.

This will allow us to see which category has the highest sales, profit, and quantity.

```python
# Group sales, profit and quantity by category
category_analysis = pd.DataFrame(superstore.groupby(['Category'])[['Sales', 'Profit', 'Quantity']].sum())
category_analysis
```

```python
# Define the "main" color palette
main_palette = ['#76A88B', '#E4B363', '#D94C4C', '#5CA4A9']

# Set for grouped plots - figure with a 2x2 grid of Axes
sns.set_palette(main_palette)
sns.set_style("whitegrid")
figure, axis = plt.subplots(1, 3, figsize=(8, 5))

# Plot barplots
cat1 = sns.barplot(x=category_analysis.index, y=category_analysis.Sales, ax=axis[0])
cat2 = sns.barplot(x=category_analysis.index, y=category_analysis.Profit, ax=axis[1])
cat3 = sns.barplot(x=category_analysis.index, y=category_analysis.Quantity, ax=axis[2])

# Set titles
cat1.set(title='Sales')
cat2.set(title='Profit')
cat3.set(title='Quantity')

# Rotate axis for x-axis
plt.setp(cat1.get_xticklabels(), rotation='vertical', size=9)
plt.setp(cat2.get_xticklabels(), rotation='vertical', size=9)
plt.setp(cat3.get_xticklabels(), rotation='vertical', size=9)

# Set spacing between subplots
figure.tight_layout()

# Show the plot
plt.show()
```

Here are our observations:

- All three categories - Furniture, Office Supplies, and Technology - have similar sales.
- Technology is the best-selling category and is also the most profitable. However, only a small quantity of technology products are sold because these products are usually one-off purchases that can last for at least 4-5 years.
- Although Furniture has similar sales to Technology, it is the least profitable and the quantity sold is also the lowest.
- Office Supplies sells the most in terms of quantity because it is a relatively cheap product.

### 2. What are the Best Selling and Most Profitable Sub-Category?

```python
# Group by sub-category
subcat_analysis = pd.DataFrame(superstore.groupby(['Sub-Category'])[['Sales', 'Profit']].sum())
```

```python
# Sort by descending order according to sales
subcat_sales = pd.DataFrame(subcat_analysis.sort_values('Sales', ascending = False))
subcat_sales
```

```python
# Sort by descending order according to profit
subcat_profit = pd.DataFrame(subcat_analysis.sort_values('Profit', ascending = False))
subcat_profit
```

```python
# Define the "main" color palette
main_palette = ['#76A88B', '#E4B363', '#BFB7AC', '#D94C4C', '#5CA4A9']

# Set for grouped plots - figure with a 1x2 grid of Axes
sns.set_style("whitegrid")
figure, axis = plt.subplots(1, 2, figsize=(12, 6))

# Plot Bar Plot for Best Selling Sub-Category
subcat1 = sns.barplot(data=subcat_sales, x=subcat_sales.index, y=subcat_sales.Sales, ax=axis[0], palette=main_palette)
subcat1.set(title="Best Selling Sub-Category")
subcat1.set_xticklabels(subcat1.get_xticklabels(), rotation="vertical", size=10)

# Plot Bar Plot for Most Profitable Sub-Category
subcat2 = sns.barplot(data=subcat_profit, x=subcat_profit.index, y=subcat_profit.Profit, ax=axis[1], palette=main_palette)
subcat2.set(title="Most Profitable Sub-Category")
subcat2.set_xticklabels(subcat2.get_xticklabels(), rotation="vertical", size=10)

# Set spacing between subplots
figure.tight_layout()

plt.show()
```

Let's examine the bar charts:

- Phones and Chairs stand out as the two most popular sub-categories in terms of sales.
- Copiers generate the highest profits, followed by Phones, Accessories, Papers, and Binders. Therefore, the marketing strategy should prioritize these products.
- Conversely, Machines, Fasteners, Supplies, Bookcases, and Tables show minimal to negative profit margins. These are items that Super Store could contemplate removing from the product catalog, adjusting their sale prices and profit margins, or negotiating for better terms with suppliers.

### 3. Which is the Top Selling Sub-Category?

```python
subcat_quantity = pd.DataFrame(superstore.groupby(['Sub-Category'])[['Quantity']].sum().sort_values('Quantity',ascending=False))
subcat_quantity
```

```python
# Define the "main" color palette
main_palette = ['#76A88B', '#E4B363', '#BFB7AC', '#D94C4C', '#5CA4A9']

# Set the color palette
sns.set_theme(style="whitegrid")

# Plot Bar Plot for Top Selling Sub-Category
sns.barplot(data=subcat_quantity, y=subcat_quantity.index, x=subcat_quantity.Quantity, palette=main_palette)
plt.title("Top Selling Sub-Category")
plt.show()
```

From this analysis, we can draw the following conclusions:

- Super Store should maintain consistently high inventory levels for the top-selling sub-categories, including Binders, Paper, Furnishings, and Phones.
- Although Copiers yield the highest profits, they have the lowest sales volume, with only 234 units sold. This can be attributed to their high price and the fact that they are typically used for several years, making it reasonable for them to have lower sales compared to other products.

### 4. Which Customer Segment is Most Profitable?

```python
segment_analysis = pd.DataFrame(superstore.groupby(['Segment'])[['Profit']].sum())
segment_analysis
```

```python
# Define the "main" color palette
main_palette = ['#76A88B', '#E4B363', '#5CA4A9']

# Set the color palette
sns.set_palette(main_palette)
sns.set_theme(style="whitegrid")

# Plot Bar Plot for Customer Segment Profitability with the specified palette
sns.barplot(data=segment_analysis, x=segment_analysis.index, y=segment_analysis.Profit, palette=main_palette)
plt.title("Customer Segment Profitability")
plt.show()
```


The consumer segment is the most profitable, followed by the corporate segment and the home office segment. Therefore, the marketing strategy should prioritize retaining and targeting customers in the consumer and corporate segments.

### 5. Which is the Preferred Ship Mode?

```python
# Define the "main" color palette
main_palette = ['#76A88B', '#E4B363', '#D94C4C', '#5CA4A9']

# Set the color palette
sns.set_palette(main_palette)
sns.set_theme(style="whitegrid")

# Get the counts of each 'Ship Mode'
ship_mode_counts = superstore['Ship Mode'].value_counts()

# Plot the counts
sns.barplot(x=ship_mode_counts.index, y=ship_mode_counts.values, palette=main_palette)
plt.title("Ship Mode")
plt.xlabel("Ship Mode")
plt.ylabel("Count")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```


By a significant margin, Standard Class stands out as the preferred and likely the most cost-effective shipping option. The other shipping methods appear to be less favored among customers and potentially more expensive.

### 6. Which Region is the Most Profitable?

```python
region_analysis = pd.DataFrame(superstore.groupby(['Region'])['Profit'].sum().reset_index())
region_analysis
```

```python
# Define the "main" color palette
main_palette = ['#76A88B', '#E4B363', '#D94C4C', '#5CA4A9']

# Plot Pie Chart
explode = [0, 0, 0, 0.1]

plt.pie(region_analysis.Profit, labels=region_analysis.Region, startangle=90, autopct="%1.0f%%", explode=explode, shadow=True, colors=main_palette)
plt.title("Most Profitable by Region")

plt.show()
```


The East and West regions are the most profitable.

### 7. Which City has the Highest Number of Sales?

```python
city_sales = pd.DataFrame(superstore.groupby(['City'])[['Sales', 'Quantity']].sum().sort_values('Sales', ascending=False))
top10 = city_sales[:10]
top10
```

```python
bottom10 = city_sales[-10:]
bottom10
```

```python
# Define the "main" color palette
main_palette = ['#76A88B', '#E4B363', '#BFB7AC', '#D94C4C', '#5CA4A9']

# Set for grouped plots - figure with a 1x2 grid of Axes
sns.set_theme(style="whitegrid")

figure, axis = plt.subplots(1, 2, figsize=(12, 5))

# Plot Bar Plot for Top 10 Cities with Highest Sales
top10c = sns.barplot(data=top10, y=top10.index, x=top10.Sales, palette=main_palette, ax=axis[0])
top10c.set_title("Top 10 Cities with Highest Sales")
top10c.set_yticklabels(top10c.get_yticklabels(), size=10)

# Plot Bar Plot for Bottom 10 Cities with Lowest Sales
bottom10c = sns.barplot(data=bottom10, y=bottom10.index, x=bottom10.Sales, palette=main_palette, ax=axis[1])
bottom10c.set_title("Bottom 10 Cities with Lowest Sales")
bottom10c.set_yticklabels(bottom10c.get_yticklabels(), size=10)

# Set spacing between subplots
figure.tight_layout()

plt.show()
```

There is a significant disparity between the cities with the highest and lowest sales. Therefore, the marketing strategy should focus on targeting the top 10 cities, where sales are the highest, to maximize revenue and business growth.

## Strategic Recommendation

Here, we revisit our business questions and share with you our high-level business recommendations.

**Business Questions**

- Which Category is Best Selling and Most Profitable?
- What are the Best Selling and Most Profitable Sub-Category?
- Which is the Top Selling Sub-Category?
- Which Customer Segment is Most Profitable?
- Which is the Preferred Ship Mode?
- Which Region is the Most Profitable?


**Recommendations**

1. **Focus on High-Performing Categories**: Concentrate your efforts on the Technology sub-category, particularly Phones and Chairs, as they are the highest selling and most profitable products. Consider bundling them with less profitable items like Bookcases and Tables to balance out the losses.

2. **Strategic Bundling**: To mitigate losses from Bookcases and Tables, consider bundling them with high-selling or high-profit sub-categories such as Chairs, Copiers, Phones, and Office Supplies. This bundling strategy can boost sales and overall profitability.

3. **Tailored Home Office Packages**: Recognize the specific needs of Home Office customers who may prioritize convenience. Create Home Office packages that include essential office products like tables, chairs, phones, copiers, storage, labels, fasteners, and bookcases. This approach can simplify their shopping experience and drive sales.

4. **Evaluate Low-Performing Products**: Assess the feasibility of retaining or discontinuing low-performing products like Supplies, Bookcases, and Tables. You may explore alternative suppliers to reduce costs or decide whether to remove them from your product catalog altogether.

5. **Target Key Customer Segments**: Recognize that Consumer and Corporate Segments constitute a significant portion of your customer base. Prioritize marketing efforts toward these segments, especially in the East and West regions, which are among the top 10 cities with the highest sales. Implement special promotions, bundles, and targeted marketing campaigns to attract and retain these customers.

By implementing these recommendations, you can optimize your product offerings, increase profitability, and effectively target your key customer segments, ultimately driving the success of Super Store.
